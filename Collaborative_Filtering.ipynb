```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns

from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
ratings_df = pd.read_csv('../ml-100k/u.data', delimiter='\t', 
                          encoding='latin-1', header=None, names=['user_id','movie_id','rating','timestamp'])
```

```python
ratings_df.head()
```

```python
movies_df = pd.read_csv('../ml-100k/u.item', sep='|', usecols=range(5), encoding='latin-1',
                     names=['movie_id', 'title', 'release_date', 'video_release_date', 'imdb_url'])
```

```python
movies_df.head()
```

```python
users_df = pd.read_csv('../ml-100k/u.user', sep='|', names=['user_id', 'age', 'sex', 'occupation', 'zip_code'],
                    encoding='latin-1', parse_dates=True) 
```

```python
users_df.head()
```

```python
movieratings_df = pd.merge(movies_df, ratings_df)
df = pd.merge(movieratings_df, users_df)
```

```python
df.head()
```

```python
# movies with high total number of ratings 

df_ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
df_ratings['total_ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
df_ratings.rename(columns = {'rating': 'avg_rating'}, inplace=True)
df_ratings.sort_values('total_ratings', ascending=False).head(10)
```

```python
# movies with highest avg ratings where total number of ratings is greater than 20

df_ratings[df_ratings['total_ratings'] > 20].sort_values('avg_rating', ascending=False).head()
```

```python
sns.jointplot(x = 'avg_rating', y = 'total_ratings', data = df_ratings )
```

```python
# removing unnecessary columns

df.drop(['user_id', 'occupation'], axis=1, inplace=True)
ratings_df.drop('timestamp', inplace = True, axis = 1) 
movies_df.drop(['video_release_date', 'imdb_url'], inplace = True, axis = 1)
```

```python
# pivot table
ratings_matrix = ratings_df.pivot_table(index=['movie_id'],columns=['user_id'],values='rating').reset_index(drop=True)
```

```python
ratings_matrix.fillna(0, inplace = True)
```

```python
ratings_matrix.head()
```

```python
similarity = 1 - pairwise_distances(ratings_matrix.values, metric="cosine")
```

```python
np.fill_diagonal(similarity, 0)
movie_similarity_df = pd.DataFrame(similarity)
```

```python
movie_similarity_df.head()
```

```python
def provide_recommendation(movie_name, top=10):
    try:
        idx = movies_df[movies_df['title']==movie_name].index.tolist()[0]
        movies_df['similarity'] = movie_similarity_df.iloc[idx]
        return movies_df.sort_values(["similarity"], ascending = False)['title'].to_list()[:top]
    except:
        print("Sorry! This movie is not available in the data! Please try another one.")   
```

```python
provide_recommendation('Speed (1994)', top=5)
```
